**Another simple way to convert the first row as a column name is to simply header=True**

```python
dff_pyspark = spark.read.csv('test.csv', header=True, inferSchema=True)
dff_pyspark.show()
```

# Adding Columns in DataFrame

```python
dff_pyspark.withColumn('Experience After Two Years',dff_pyspark['Experience']+2)
```

```python
dff_pyspark.withColumn('Experience After Two Years',dff_pyspark['Experience']+2).show()
```

# Drop Columns

```python
dff_pyspark.drop('Experience After Two Years').show()
```

```python
dff_pyspark.show()
```
